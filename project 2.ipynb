```python
# ===== IMPORTS =====
import numpy as np
import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer

# ===== DATAFRAME =====
df = pd.DataFrame({
    "text": [
        "Meditation and yoga can improve mental health",
        "Fruits, whole grains and vegetables helps control blood pressure",
        "These are the latest fashion trends for this week",
        "Vibrant color jeans for male are becoming a trend",
        "The concert starts at 7 PM tonight",
        "Navaratri dandiya program at Expo center in Mumbai this october",
        "Exciting vacation destinations for your next trip",
        "Maldives and Srilanka are gaining popularity in terms of low budget vacation places"
    ]
})

# ===== ENCODER =====
encoder = SentenceTransformer("all-mpnet-base-v2")

# ===== EMBEDDINGS =====
vectors = encoder.encode(df["text"])
vectors = np.array(vectors).astype("float32")

# ===== FAISS INDEX =====
dim = vectors.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(vectors)

# ===== SEARCH QUERY =====
query = "health related topics"
new_vec = encoder.encode([query])
new_vec = np.array(new_vec).astype("float32")

distances, I = index.search(new_vec, k=2)

print("Distances:", distances)
print("Results:")
for idx in I[0]:
    print("-", df.iloc[idx]["text"])
```
